```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
cars = sns.load_dataset('mpg')
```

```python
plt.hist(x= cars.mpg, label = 'mpg')
```

Shows the shape of the data clearly but missing information

```python
plt.plot(cars.mpg, label = 'mpg')
```

Shows the individual datapoints for each car, but the lines make it difficult to interpret

```python
plt.pie(cars.mpg)
plt.show()
```

It looks pretty, but this seems completely useless other than as a basis for some sort of pseudo-random number generator

```python
sns.boxplot(cars.mpg)
```

Missing information, and does not specificy the quantity of datapoints, but this displays the shape of the data effectively

```python
sns.lmplot(x='weight', y='horsepower', data=cars)
```

Good plot, shows a strong correlation between weight and horsepower

```python
sns.catplot(x='weight', y='horsepower', data=cars, kind = 'bar', palette="pastel")
```

This also shows the correlation but it is not as digestible as the previous plot. 

```python
sns.boxplot(x='weight', y='horsepower', data=cars, palette="pastel")
```

This one also shows the correlation but aside from that it is flawed, the values on the x axis are not being displayed

```python
sns.boxplot(x='origin', y='mpg', data=cars, palette="pastel")
```

This plot seems to have no obvious downside. It tells a clear story: the cars' origins seem to strongly correlate with their mpg (or lack thereof)
The y-axis is cut off at 10, but it does not disproportionately skew the shape of the data, so it is forgiveable.

```python
sns.lmplot(y='mpg', x='origin', data=cars, fit_reg=False)
```

This plot tells the same story as the previous plot, but it is much less digestible 

```python
sns.pointplot(x='origin', y='mpg', data=cars, palette="pastel")
```

Both the best and worst part of this plot is that the y-axis cutoff makes it appear as if the USA is far below where it actually is. Framing visualizations this way is extremely common when the person creating the graph wants the viewer to draw a certain conclusion. The fact that the y-axis starts at 20, and japan and europe are far above, affects the perception of a reader or viewer differently from the initial boxplot, which started at 10. This technique can be used to mislead people, as it usually artificially stresses how far above or below the average certain categories' data is. This seems to be extremely common in graphs that get shown on television, in magazines, online, etc. For example, a weather program might display a miniscule temperature difference in the form of two bars, one bar only reaching the middle of the screen, and the other bar reaching the top, making it seem as if the change is much more drastic than it actually is.

```python
sns.catplot(x='origin', y='mpg', data=cars, palette="pastel")
```

Compared to the previous graph, this graph is much more honest. For those interested in portraying the data holistically, then this is preferable to the previous graph. But if all one wanted was to help the viewer conlcude that the usa is lagging behind reagarding mpg, the first graph is much more efficient at conveying that message. 

```python
sns.catplot(x='origin', y='mpg', data=cars, kind='bar', palette="pastel", ci=95)
```

This takes the best elements of the previous two graphs and combines them. It has the visual appeal of the misleading plot, as well as the honesty of the second plot. Cutting off the y-axis at 20 was simply unnecessary, unless the goal is to create more shock value. This graph clearly shows that it is not necessary to do any sort of manipulation to make the USA look bad. Even in this fairer visualization, the USA is clearly in last place

```python
g = sns.lmplot(x='origin', y='mpg', data=cars, fit_reg=False)
g.set_ylabels("MPG")
g.set_xlabels("Origin")
plt.title('MPG by Origin')
plt.show()
```
