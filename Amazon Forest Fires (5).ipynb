The following data concerns forest fires in the Amazon rainforest. The data comes from the Brazilian government's statistics, and spans from 1997-2017. There is data for 22 of Brazil's states as well as the Distrito Federal. Presumably, these 23 territories comprise all the parts of Brazil that contain sections of the Amazon. It has been a notable topic in recent years that there are severe issues with fires in the Amazon rainforest, but what story do the statistics say? We would expect to see a steady increase over the past 10 years in accordance with the news coverage. The data being separated by state will allow us to analyze any unique data and possibly explain outliers.

This analysis of 20 years of forest fires in the Amazon has three target areas:

1. How has the number of fires in Brazil as a whole changed from 1998 to 2017? Does ignoring the changes of the individual  datapoints, only considering the overall average number, yield a different conclusion from the one reached by analyzing individual states?

2. Which states have seen the most significant changes? Which states have seen less change? Do the numbers tend to increase or decrease over time? Are there a few states that are outliers and skewing any overall impressions of the data?

3. Why does the data for Rio have three values for each date? Are there significant differences between the datasets that could provide insight (e.g. one set has abnormally high values, one has low values)? Are there any other similar inconsistencies in the data as a whole, outside of Rio's data? What are the most likely explanations? What are some other possible explanations that are less plausible, but could explain the issue? Finally, how could this issue be resolved? Does it need to be resolved, or is the data largely unaffected by the discrepancies? Potential solutions could include taking the mean values for each of the three sets, or choosing the set that is most congruent with the rest of the data from other states.

Rio has three duplicate data sets. For each month/year combination, there are three numbers recorded. This aspect is unique to Rio's data: every other state (plus the Distrito Federal) has only one set of data. There are several possible reasons for this inconsistency. The simplest reason is that there are three conflicting sources that provide different numbers. Similarly, there could be different criteria for each of the three sets of data of Rio, regardless of whether the different sets come from the same source or not. For example, a small campfire that does not spread very far might be counted in one set, but ommitted from another. Another, less likely possibility, is that there are not actually three conflicting sets, but three datapoints for each month, possibly for the first ten days/next ten days/final days of the month. However, this is very unlikely. The data would have presumably been organized to show the three 'identical' datapoints consecutively, as opposed to being formatted the same way as the datasets for other states. Another possibility is that the data describes three different geographical sections of Rio. A final possibility is that the dataset is inaccurately describing all three of these sets as being Rio's data, when in actuality only one set is for Rio, while the other two sets are for different states. This final possibility seems the least likely, as the data appears to be carefully curated, and is taken directly from a government source. However, it would explain the discrepancy.

On average, the fires increased in frequency by 23 percent from 1997-2007 to 2008-2017. The median increase was about 18.7%, suggesting that while the mean is realtively close to the center, there are some outlier values that at pulling it away. The two locations that have seen the most drastic increases are the Distrito Federal (91.8%) and Goias (75.0%). Sao Paulo saw the largest reduction (31.4%) which stands out compared to 17.0%, the next highest reduction. Overall, these outliers are enough to slightly skew the mean but not so drastically that it dilutes the effectiveness of the summary statistics. Even with the two highest percentages removed, the trend is that the number of fires increased.
