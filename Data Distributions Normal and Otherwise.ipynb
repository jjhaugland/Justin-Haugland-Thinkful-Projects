```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from matplotlib.pyplot import hist
wave = np.random.rayleigh([1], 100)
mean = np.mean(wave)
sd = np.std(wave)
plt.hist(wave)
plt.axvline(wave.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(wave.mean() + wave.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(wave.mean()-wave.std(), color='g', linestyle='dashed', linewidth=2) 
plt.show()
#mean and standard deviation are not ideal as there is skew. 
#the level of suitability depends on the precision desired
#for a rough estimate of the spread and center of values, the mean and stdev are somewhat useful
#however, the skew would not be reflected by those numbers
#these summary stats imply/assume/work best with a normal or normal-esque distribution
```

```python
coin = np.random.geometric([0.5], 100)
mean = np.mean(coin)
sd = np.std(coin)
plt.hist(coin, bins = 15)
plt.axvline(coin.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(coin.mean() + coin.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(coin.mean()-coin.std(), color='g', linestyle='dashed', linewidth=2) 
plt.show()
#plot is skewed, so the summary statistics are not as relevant as in other distributions
```

```python
log = np.random.logistic([0, 1], 100)
mean = np.mean(log)
sd = np.std(log)
plt.hist(log, bins=5)
plt.axvline(log.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(log.mean() + log.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(log.mean()-log.std(), color='g', linestyle='dashed', linewidth=2) 
plt.show()
#mean shows the center, but in this instance the summary statistics do not appear to be useful or relevant, despite being accurate
```

```python
n, p = 10, .5
successes = np.random.binomial(n, p, 100)
mean = np.mean(successes)
sd = np.std(successes)
plt.hist(successes, bins = 10)
plt.axvline(successes.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(successes.mean() + successes.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(successes.mean()-successes.std(), color='g', linestyle='dashed', linewidth=2) 
plt.show()
#mean and standard deviation are slighlty useful measures of the center and spread
#as the number of trials increases, the distribution becomes more symmetrical, and the summary statistcs estimate the center more accurately
```

```python
loc, scale = 0, 1
var = np.random.laplace(loc, scale, 100)
mean = np.mean(var)
sd = np.std(var)
plt.hist(var)
plt.axvline(var.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(var.mean() + var.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(var.mean()-var.std(), color='g', linestyle='dashed', linewidth=2) 
plt.show()
#mean and stdev seem to be relatively good descriptors of the data set
#histogram appears to resemble a normal distribution
```

```python
lam = 5
fish = np.random.poisson(lam, 100)
mean = np.mean(fish)
sd = np.std(fish)
plt.hist(fish)
plt.axvline(fish.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(fish.mean() + fish.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(fish.mean()-fish.std(), color='g', linestyle='dashed', linewidth=2) 
plt.show()
#Mean and standard deviation are somewhat descriptive of the data set, although there is skew
```

```python
#The distributions that are most accurately describe by mean and standard deviation have commonalities
#They have a peak in the center, and are relatively symmetrical
#This shape is reminiscent of normal distributions

#Data that is assymetrical (geometric) or data that lacks a peak in the center (logistic) is not suited to being described with mean and standard deviation
```

```python
var1 = np.random.normal (5, .5, 100)
var2 = np.random.normal (10, 1, 100)
var3 = var1 + var2
mean = np.mean(var3)
sd = np.std(var3)
plt.hist(var3)
plt.axvline(var3.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(var3.mean() + var3.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(var3.mean()-var3.std(), color='g', linestyle='dashed', linewidth=2) 
plt.show()
```

```python
#the histogram is noticeably skewed, but descriptive statistics are still applicable to a certain extent
#theoretically, the distribution should be symmetrical. 100 trials is relatively low
#with 1000 trials, the distribution is much closer to being perfectly symmetrical
```

```python
var1 = np.random.normal (5, .5, 1000)
var2 = np.random.normal (10, 1, 1000)
var3 = var1 + var2
mean = np.mean(var3)
sd = np.std(var3)
plt.hist(var3)
plt.axvline(var3.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(var3.mean() + var3.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(var3.mean()-var3.std(), color='g', linestyle='dashed', linewidth=2) 
plt.show()
```
